```python
import pandas as pd
df = pd.read_excel('sample_data.xlsx')
df[:]
```

```python
#Assignment: Filtering Data with query()
#Fetch all records where the education is 'bachelors' and the gender is 'female'.

df.query('education=="bachelors" and gender=="female"')
```

```python
#Assignment: Creating a New Column with apply()
#Create a new column named income_level. Use the apply() function on the inome column. If inome is '<=50k',
#the income_level should be 'low_income'; otherwise, it should be 'high_income'.
#df['inome'].unique()
df['income_level'] = df['inome'].apply(lambda inome:'low_income' if inome=='<=50k' else 'high_income')
df[:20]
```

```python
#Assignment: Filtering Data with query() - Part 2
#Find all records where the job is 'private' and marrital_status is 'married-civ-spouse'.
df.query('job=="private" and marrital_status=="married-civ-spouse"')
```

```python
#Assignment: Exploring Unique Values
#Find all unique country values in the dataset
df['country'].unique()
```

```python
#Assignment: Filtering by Numerical Condition
#Select all rows where age is greater than 60.
df.query('age>60')
```

```python
#Assignment: Creating a Column with apply() based on capital_gain
#Create a new column called has_capital_gain. If captial_gain is greater than 0, set has_capital_gain to 1,
# otherwise set it to 0.
#df['captial_gain'].unique()
df['has_capital_gain'] = df['captial_gain'].apply(lambda captial_gain: '1' if captial_gain > 0 else '0')
df[:2]
```

```python
#Assignment: Further Cleaning - Handling '?' values
#Identify all columns that contain '?' values and replace these '?' with 'unknown'.
df = df.replace('?','unknown')
df[:]
```
